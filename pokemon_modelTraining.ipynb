```python
# -*- coding: utf-8 -*-
"""
Created on Thu Apr 11 15:23:12 2019
@author: rjlin
"""

from keras.models import Sequential  #用來啟動 NN
from keras.layers import Conv2D  # Convolution Operation
from keras.layers import MaxPooling2D # Pooling
from keras.layers import MaxPool2D
from keras.layers import Flatten
from keras.layers import Dense # Fully Connected Networks
from keras.layers import Activation
from keras.preprocessing.image import ImageDataGenerator
from keras.utils import np_utils

import numpy as np
import pandas as pd
import random


google_gen = ImageDataGenerator(
    featurewise_center=True,
    featurewise_std_normalization=True,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    
    zoom_range=0.2, # 隨機縮放的強度
    horizontal_flip=True
)


model = Sequential()  
model.add(Conv2D(8, (5, 5), padding='same', input_shape=(3, 100, 100))) # layer 1
model.add(Activation('relu'))
model.add(MaxPool2D(pool_size=(2,2))) # max-pooling
model.add(Conv2D(16, (5, 5), padding='same')) # layer 2
model.add(Activation('relu'))
model.add(MaxPool2D(pool_size=(2,2))) # max-pooling
model.add(Conv2D(32, (5, 5), padding='same'))
model.add(Activation('relu')) # layer 3
model.add(MaxPool2D(pool_size=(2,2))) # max-pooling
model.add(Conv2D(64, (5, 5), padding='same'))
model.add(Activation('relu')) # layer 4
model.add(MaxPool2D(pool_size=(2,2))) # max-pooling
model.add(Conv2D(128, (5, 5), padding='same'))
model.add(Activation('relu')) # layer 5
model.add(MaxPool2D(pool_size=(2,2))) # max-pooling
model.add(Conv2D(128, (5, 5), padding='same'))
model.add(Activation('relu')) # layer 6
model.add(MaxPool2D(pool_size=(2,2))) # max-pooling

model.add(Flatten())

model.add(Dense(output_dim = 200, activation = 'relu'))
model.add(Dense(output_dim = 806, activation = 'sigmoid'))

model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
model.summary()

model.fit_generator(generator=google_gen.flow_from_directory('/Users/yan/Documents/ML_Pokemon/png_train', target_size=(100, 100), color_mode='rgb'), 
                    validation_data=google_gen.flow_from_directory('/Users/yan/Documents/ML_Pokemon/png_test', target_size=(100, 100), color_mode='rgb'), 
                    steps_per_epoch=3000,
                    validation_steps=300, 
                    epochs = 20
                    )

model.save('pokemon_classifer.h5')
```
